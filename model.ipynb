```python
from sklearn import linear_model
from sklearn.neighbors import KNeighborsRegressor
from sklearn import ensemble
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
import pickle
import pandas as pd
```

```python
df = pd.read_csv("shuffled_data.csv")

tweets = list(df["tweet"])
labels = list(df["class"])

df.head()
```

```python
tweets[0]
```

```python
def process_text(texts):
    new_texts = []
    for text in texts:
        text = text.lower()
        # special characters
        text = re.sub(r'\W', ' ', text)
        # single characters
        text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)
        # substitute multiple spaces with single one
        text = re.sub(r'\s+', ' ', text)
        # single characters from the start
        text = re.sub(r'\^[a-zA-Z]\s+', ' ', text)
        # lemmatization
        text = text.split(' ')
        text = [stemmer.lemmatize(word) for word in text]
        new_texts.append(' '.join(text))
    return new_texts
```

```python
from nltk.stem import WordNetLemmatizer

stemmer = WordNetLemmatizer()
tweets_processed = process_text(tweets)
```

```python
print(tweets_processed[0], '\n', tweets[0])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer(stop_words=stopwords.words('english'))
X = vect.fit_transform(tweets_processed).toarray()
```

```python
tweets_processed
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfd = TfidfTransformer()
X = tfd.fit_transform(X).toarray()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
from sklearn import metrics
```

```python
fits = model.fit(x_train, y_train)
```

```python
from sklearn import metrics
score = metrics.r2_score(y_test, fits.predict(x_test))
mse = metrics.mean_squared_error(y_test, fits.predict(x_test))
mae = metrics.mean_absolute_error(y_test, fits.predict(x_test))
```
